```python
import pandas as pd

# 파일 경로

data = '240307_1237_4line_I2C_pointingv20'
filename = 'myData/'+data+'.csv'
# 파일 로드
data = pd.read_csv(filename)
# 데이터의 처음 몇 줄을 확인
data.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# contact가 1인 데이터만 필터링
contact_data = data[data['contact'] == 1]

# theta 계산 (atan2는 범위가 -pi에서 pi이므로 적합)
theta = np.arctan2(contact_data['sin'], contact_data['cos'])

# z 값을 가로축으로, theta를 세로축으로 하는 그래프 다시 그리기
plt.figure(figsize=(10, 6))
plt.scatter(contact_data['z'], theta, color='red', alpha=0.5)
plt.ylabel('Theta (radians)')
plt.xlabel('Z value')
plt.title('Graph of Z value vs Theta for Contact=1')
plt.grid(True)
plt.show()

```
