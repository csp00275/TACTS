```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for the hammer
m_hammer = 0.5  # mass of the hammer in kg
R = 0.2  # length of the hammer in m
g = 9.81  # acceleration due to gravity in m/s^2

# Calculate the velocity of the hammer at the bottom
v_hammer = np.sqrt(2 * g * R)

# Constants for the balloon
p = 1* 10**3  # pressure inside the balloon in Pa
r_balloon = 0.2  # radius of the balloon in m
h_balloon = 0.16

# Calculate the elasticity of the balloon
k = p / ( np.pi * r_balloon**2 * h_balloon)

# Calculate the initial deformation of the balloon
x0 = m_hammer * v_hammer / k

# Time array
t = np.linspace(0, 2*np.pi/np.sqrt(k/m_hammer), 1000)

# Calculate the acceleration of the object
a = -k/m_hammer * x0 * np.cos(np.sqrt(k/m_hammer) * t)

# Calculate the force on the object
F = m_hammer * a

# Plot the force on the object over time
plt.figure(figsize=(10, 6))
plt.plot(t, F, label=f'm_hammer = {m_hammer} kg, p = {p} Pa')
plt.title('Force on object over time')
plt.xlabel('Time (s)')
plt.ylabel('Force (N)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
R = 0.2  # length of the hammer in m
g = 9.81  # acceleration due to gravity in m/s^2
r_balloon = 0.2  # radius of the balloon in m
h_balloon = 0.16

# Array of hammer masses and pressures
m_ham_values = np.linspace(0.1, 0.5, 5)
p_values = np.linspace(1, 5, 5)  # pressures in kPa

# Convert pressures to Pa for calculations
p_values_pa = p_values * 10**3

# Initialize arrays to store maximum force and period
max_force = np.zeros((len(m_ham_values), len(p_values_pa)))
period = np.zeros((len(m_ham_values), len(p_values_pa)))

# Calculate maximum force and period for each combination of hammer mass and pressure
for i, m_ham in enumerate(m_ham_values):
    for j, p in enumerate(p_values_pa):
        # Calculate the velocity of the hammer at the bottom
        v_hammer = np.sqrt(2 * g * R)
        # Calculate the elasticity of the balloon
        k = p / ( np.pi * r_balloon**2 * h_balloon)
        # Calculate the initial deformation of the balloon
        x0 = m_ham * v_hammer / k
        # Calculate the maximum force
        max_force[i, j] = m_ham * k * x0
        # Calculate the period
        period[i, j] = 2 * np.pi / np.sqrt(k / m_ham)

# Create 3D plots
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(m_ham_values, p_values, max_force)
ax1.set_xlabel('Hammer mass (kg)')
ax1.set_ylabel('Pressure (kPa)')
ax1.set_zlabel('Max force (N)')
ax1.set_title('Max force for different hammer masses and pressures')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(m_ham_values, p_values, period)
ax2.set_xlabel('Hammer mass (kg)')
ax2.set_ylabel('Pressure (kPa)')
ax2.set_zlabel('Period (s)')
ax2.set_title('Period for different hammer masses and pressures')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
R = 0.2  # length of the hammer in m
g = 9.81  # acceleration due to gravity in m/s^2
r_balloon = 0.2  # radius of the balloon in m
h_balloon = 0.16

# Array of hammer masses and pressures
m_ham_values = np.linspace(0.1, 0.5, 100)
p_values = np.linspace(1000, 5000, 100)  # pressures in Pa

# Initialize arrays to store maximum force and period
max_force_m_ham = np.zeros(len(m_ham_values))
period_m_ham = np.zeros(len(m_ham_values))
max_force_p = np.zeros(len(p_values))
period_p = np.zeros(len(p_values))

# Calculate maximum force and period for different hammer masses
for i, m_ham in enumerate(m_ham_values):
    # Calculate the velocity of the hammer at the bottom
    v_hammer = np.sqrt(2 * g * R)
    # Calculate the elasticity of the balloon
    k = p_values[-1] / (np.pi * r_balloon**2 * h_balloon)
    # Calculate the initial deformation of the balloon
    x0 = m_ham * v_hammer / k
    # Calculate the maximum force
    max_force_m_ham[i] = m_ham * k * x0
    # Calculate the period
    period_m_ham[i] = 2 * np.pi / np.sqrt(k / m_ham)

# Calculate maximum force and period for different pressures
for i, p in enumerate(p_values):
    # Calculate the elasticity of the balloon
    k = p / (np.pi * r_balloon**2 * h_balloon)
    # Calculate the initial deformation of the balloon
    x0 = m_ham_values[-1] * v_hammer / k
    # Calculate the maximum force
    max_force_p[i] = m_ham_values[-1] * k * x0
    # Calculate the period
    period_p[i] = 2 * np.pi / np.sqrt(k / m_ham_values[-1])

# Plot maximum force vs hammer mass
plt.figure(figsize=(10, 6))
plt.plot(m_ham_values, max_force_m_ham)
plt.xlabel('Hammer mass (kg)')
plt.ylabel('Max force (N)')
plt.title('Maximum Force vs Hammer Mass')
plt.grid(True)
plt.show()

# Plot period vs hammer mass
plt.figure(figsize=(10, 6))
plt.plot(m_ham_values, period_m_ham)
plt.xlabel('Hammer mass (kg)')
plt.ylabel('Period (s)')
plt.title('Period vs Hammer Mass')
plt.grid(True)
plt.show()

# Plot maximum force vs pressure
plt.figure(figsize=(10, 6))
plt.plot(p_values, max_force_p)
plt.xlabel('Pressure (Pa)')
plt.ylabel('Max force (N)')
plt.title('Maximum Force vs Pressure')
plt.grid(True)
plt.show()

# Plot period vs pressure
plt.figure(figsize=(10, 6))
plt.plot(p_values, period_p)
plt.xlabel('Pressure (Pa)')
plt.ylabel('Period (s)')
plt.title('Period vs Pressure')
plt.grid(True)
plt.show()
```
