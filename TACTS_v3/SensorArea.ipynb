```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def plot_graph(D, d, h, nr, ny, name, pattern='even',show_labels=True):
    
    R = D / 2
    r = d / 2
    L = np.pi * D
    FoV = 25
    FoV = np.deg2rad(FoV)
    
    
    CoV = 2*np.arcsin((R-r)/R*np.tan(FoV/2))
        
    plt.rcParams['font.family'] = 'Malgun Gothic'
    plt.style.use('default')
    plt.rcParams['figure.figsize'] = (L/100, h/100)
    plt.rcParams['font.size'] = 12
    
    ns = nr * ny

    cp_x = L - (L / (2 * nr)) #   
    op_x = L / nr #  
    cp_y = h - (h / (2 * ny))  #   
    op_y = h / ny  #  

    fig, ax = plt.subplots()

    ax.set_xlim(0, L)
    ax.set_ylim(0, h)
    ax.set_xlabel('Circumference')
    ax.set_ylabel('Height')
    
    # x  
    x_ticks = np.arange(0, L+ L / 4, step= L / 4)
    plt.xticks(x_ticks, [f'{int(x / (np.pi))} ' for x in x_ticks])

    # y  : 0 L 40 
    plt.yticks(np.arange(0, h + h/4, step=h/4))

    ax.add_patch(
        patches.Rectangle(
            (0, 0), L, h, 
            edgecolor='lightblue',
            facecolor='lightblue',
            fill=True,
        ))
    
    #   
    major_axis = R*CoV
    minor_axis = (R - r) * 2 * np.tan(FoV / 2)
    
    offset = op_x / 2 if pattern == 'odd' else 0

    for j in range(0, ny):
        if pattern == 'odd' and j % 2 == 1:
            local_nr = nr + 1
        else:
            local_nr = nr

        for i in range(0, local_nr):
            local_offset = offset if j % 2 == 1 else 0
            ax.add_patch(
                patches.Ellipse(
                    (cp_x - i * op_x + local_offset, cp_y - j * op_y),
                    major_axis, minor_axis, # , 
                    edgecolor='black',
                    facecolor='red',
                    fill=True,
                    linewidth=1
                ))

    plt.title('TACTS_' +name+ '_r{0}_z{1}'.format(nr, ny))
    
    
    if not show_labels:
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        #ax.set_xlabel('')
        #ax.set_ylabel('')
        plt.title('')
    
    plt.show()

    print('  : {0}'.format(ns))
    print('major_axis: ', major_axis)
    print('minor_axis: ',minor_axis)

#   
plot_graph(200, 100, 160, 8, 8,'Axial', pattern='odd') # D, d, nr, nz
```

```python
def plot_graph(tri_height, frame_width=0.6, frame_height=0.3, angle_1=25, angle_2=35):
    rect_width = 0.44
    rect_height = 0.1
    rect_x = (frame_width - rect_width) / 2
    rect_y = 0
    offset = 0.0665

    #      
    tri_base_1 = 2 * tri_height * np.tan(np.deg2rad(angle_1 / 2)) / 10
    tri_base_2 = 2 * tri_height * np.tan(np.deg2rad(angle_2 / 2)) / 10

    plt.figure(figsize=(frame_width * 10, frame_height * 10))

    plt.gca().add_patch(patches.Rectangle((rect_x, rect_y), rect_width, rect_height, edgecolor='black', facecolor='black'))

    #     
    plt.gca().add_patch(patches.Polygon([(rect_x + offset, rect_y + rect_height), (rect_x + offset + tri_base_1 / 2, rect_y + rect_height + tri_height / 10), (rect_x + offset - tri_base_1 / 2, rect_y + rect_height + tri_height / 10)], closed=True, color='red', alpha=0.5))
    plt.gca().add_patch(patches.Polygon([(rect_x + rect_width - offset, rect_y + rect_height), (rect_x + rect_width - offset + tri_base_2 / 2, rect_y + rect_height + tri_height / 10), (rect_x + rect_width - offset - tri_base_2 / 2, rect_y + rect_height + tri_height / 10)], closed=True, color='blue', alpha=0.5))

    plt.xlim(-frame_width*10, frame_width*10)
    plt.ylim(0, frame_height*10)
    plt.axis('off')
    plt.title('Frame with Rectangle and Isosceles Triangles')
    plt.show()

plot_graph(50)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def plot_graph_with_circles(D, d, h, nr, ny, name, pattern='even', show_labels=True, n_x_dots=0, n_y_dots=0, circle_color='black'):
    R = D / 2
    r = d / 2
    L = np.pi * D
    FoV = 25
    FoV = np.deg2rad(FoV)
    CoV = 2 * np.arcsin((R-r) / R * np.tan(FoV/2))

    plt.rcParams['font.family'] = 'Malgun Gothic'
    plt.style.use('default')
    plt.rcParams['figure.figsize'] = (L/50, h/50)
    plt.rcParams['font.size'] = 12

    ns = nr * ny

    cp_x = L - (L / (2 * nr)) #   
    op_x = L / nr #  
    cp_y = h - (h / (2 * ny))  #   
    op_y = h / ny  #  

    fig, ax = plt.subplots()

    ax.set_xlim(0, L)
    ax.set_ylim(0, h)
    ax.set_xlabel('Circumference')
    ax.set_ylabel('Height')

    # x  
    x_ticks = np.arange(0, L + L / 4, step= L / 4)
    plt.xticks(x_ticks, [f'{int(x / (np.pi))} ' for x in x_ticks])

    # y  
    plt.yticks(np.arange(0, h + h/4, step=h/4))

    #  
    ax.add_patch(
        patches.Rectangle(
            (0, 0), L, h, 
            edgecolor='gainsboro',
            facecolor='gainsboro',
            fill=True,
        ))
    
    #   
    major_axis = R * CoV
    minor_axis = (R - r) * 2 * np.tan(FoV / 2)

    #  
    offset = op_x / 2 if pattern == 'odd' else 0
    for j in range(0, ny):
        if pattern == 'odd' and j % 2 == 1:
            local_nr = nr + 1
        else:
            local_nr = nr

        for i in range(0, local_nr):
            local_offset = offset if j % 2 == 1 else 0
            ax.add_patch(
                patches.Ellipse(
                    (cp_x - i * op_x + local_offset, cp_y - j * op_y),
                    major_axis, minor_axis, # , 
                    #edgecolor='black',
                    facecolor='blue',
                    fill=True,
                    linewidth=1
                ))

    #   
    if n_x_dots > 0 and n_y_dots > 0:
        y_dots = np.linspace(0, h, n_y_dots)
        x_dots = np.linspace(0, L, n_x_dots+1)
        circle_radius = 0.5 #   
        for y in y_dots:
            for x in x_dots:
                ax.add_patch(
                    patches.Circle(
                        (x, y),
                        circle_radius,
                        color=circle_color, #      
                        fill=True
                    ))

    plt.title('TACTS_' + name + '_r{0}_z{1}'.format(nr, ny))
    
    if not show_labels:
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_xticklabels([])
        ax.set_yticklabels([])
    
    plt.show()

    print('  : {0}'.format(ns))
    print('major_axis: ', major_axis)
    print('minor_axis: ', minor_axis)

#   ( 18,  2   )
plot_graph_with_circles(210, 100, 160, 12, 4, 'previous', pattern='even', show_labels=False, n_x_dots=72, n_y_dots=21, circle_color='yellow')
```
